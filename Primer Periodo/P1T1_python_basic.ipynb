##Ejercicio 3: Vector/lista

Permita ingresar en una lista/vector que sea dinamica. Si le doy un numero, me pida ese numero de datos... Ordenarla de menor a mayor

```python
total_numeros = int(input("Ingrese el numero total de numeros "))
my_list = []
for i in range(total_numeros):
  my_list.append(input(f"Ingrese el numero {i+1}: "))
my_list.sort()
print(my_list)
```
